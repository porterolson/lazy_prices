```python
def vectorizer_sbert(doc1, doc2):
    #first take in the two documents
    doc1_chunks=chunk_item1a(doc1) #chunk both of them
    doc2_chunks=chunk_item1a(doc2)


    #throw all the chunks thru sbert
    embs_1 = model.encode(doc1_chunks, batch_size=32, normalize_embeddings=True)

    #get the row-wise mean, because embs_1 will be a matrix where the embedding for a given chunk will be that cols vector (i.e. embedding for chunk 1 is col 1)
    doc1_vec = embs_1.mean(axis=0)
    doc1_vec = doc1_vec / np.linalg.norm(doc1_vec) # renormalize the vector

    #throw all the chunks thru sbert
    embs_2 = model.encode(doc2_chunks, batch_size=32, normalize_embeddings=True)

    #get the row-wise mean, because embs_1 will be a matrix where the embedding for a given chunk will be that cols vector (i.e. embedding for chunk 1 is col 1)
    doc2_vec = embs_2.mean(axis=0)
    doc2_vec = doc2_vec / np.linalg.norm(doc2_vec) # renormalize the vector


    #return both of the vectors
    return doc1_vec,doc2_vec
```

```python
#chats
import numpy as np
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2", device="cuda")  # if you have GPU

tokenizer = model.tokenizer

def chunk_item1a(text, chunk_tokens=200, stride=75):
    if text is None:
        return []
    text = str(text).strip()
    if not text:
        return []

    ids = tokenizer(text, add_special_tokens=False)["input_ids"]
    step = chunk_tokens - stride

    chunks = []
    for start in range(0, len(ids), step):
        chunk_ids = ids[start:start+chunk_tokens]
        if not chunk_ids:
            break
        chunks.append(tokenizer.decode(chunk_ids))
        if start + chunk_tokens >= len(ids):
            break
    return chunks

def doc_to_vec(doc, batch_size=64):
    chunks = chunk_item1a(doc)
    if len(chunks) == 0:
        return None
    embs = model.encode(chunks, batch_size=batch_size, normalize_embeddings=True)
    v = embs.mean(axis=0)
    v = v / np.linalg.norm(v)
    return v

# Cache: (cik, year) -> vector
vec_cache = {}

# Precompute once
for row in tqdm(data.select(["cik", "year", ITEM]).iter_rows(named=True)):
    key = (row["cik"], row["year"])
    doc = row[ITEM]
    if doc is None or str(doc).strip() == "":
        vec_cache[key] = None
    else:
        vec_cache[key] = doc_to_vec(doc)
```

```python
#there is most definitely a faster way to do this, but I am not a CS major so this works for me haha

cik_list=data['cik'].unique().to_list()
cik_list.sort() #sort it so we match how our data df is

# cik_list = cik_list[:1] #first three for testign


sim_list_cos=[]
sim_list_pearson=[]
distance_list=[]
manhattan_list=[]
rbf_list=[]
poly_list=[]

for cik in tqdm(cik_list): #this method assumes that my data is sorted on cik and year where cik is increasing and year is decreasing (within cik)
    sub=data.filter(pl.col('cik')==cik) #filter to cik
    year_item=sub.select([ITEM,'year']) #filter to item
    year_list=year_item['year'].to_list() #get a list of the years so we can iterate on it too


    sim_list_cos_cik=[]
    sim_list_pearson_cik=[]
    distance_list_cik=[]
    manhattan_list_cik=[]
    rbf_list_cik=[]
    poly_list_cik=[]

    for year in year_list:
        try:
            doc1 = year_item.filter(pl.col('year') == year).select(ITEM).item() #get this year
            doc2 = year_item.filter(pl.col('year') == (year - 1)).select(ITEM).item() #last years item

        except ValueError:
            sim_list_pearson_cik.append(None) #when compariing our earliest year to the year before it throws an error, so we append a none (null) value
            distance_list_cik.append(None)
            sim_list_cos_cik.append(None)
            manhattan_list_cik.append(None)
            rbf_list_cik.append(None)
            poly_list_cik.append(None)
            continue

        doc_list=[doc1,doc2]
        try:
            # X = vectorizer.fit_transform(doc_list) #vectorize it (uncomment this if using tfidf and count)

            # #convert sparse rows to dense arrays
            # v1 = X[0].toarray().ravel() #get the first vec from the vectorizer
            # v2 = X[1].toarray().ravel() #second one
            # handle nulls / empties explicitly

            v1 = vec_cache.get((cik, year))
            v2 = vec_cache.get((cik, year-1))
            if v1 is None or v2 is None:
                sim_list_pearson_cik.append(None) #when compariing our earliest year to the year before it throws an error, so we append a none (null) value
                distance_list_cik.append(None)
                sim_list_cos_cik.append(None)
                manhattan_list_cik.append(None)
                rbf_list_cik.append(None)
                poly_list_cik.append(None)
                continue
            else:
                cos_similarity = float(np.dot(v1, v2))


                doc1 = str(doc1).strip()
                doc2 = str(doc2).strip()
                if doc1 == "" or doc2 == "":
                    sim_list_pearson_cik.append(None)
                    distance_list_cik.append(None)
                    sim_list_cos_cik.append(None)
                    manhattan_list_cik.append(None)
                    rbf_list_cik.append(None)
                    poly_list_cik.append(None)
                    continue



                cos_similarity = float(np.dot(v1, v2)) # compute the cos similarity for each to the previous year
                similarity_pearson = np.corrcoef(v1, v2)[0, 1] #pearson correlation
                distance = np.linalg.norm(v1 - v2) #euclidean distance

                X = np.vstack([v1, v2])   # shape = (2, 768)

                manhattan = manhattan_distances(X)[0, 1] #manhattan dist

                gamma = 1.0 / 768 #gamma, we choose this param, this is a common choice for gamma without doing CV
                rbf_sim = rbf_kernel(X, gamma=gamma)[0, 1] #rbf kernel similiarity

                ######################## POLYNOMIAL PARAMS
                degree = 2
                gamma_poly = 1.0 / 768 #768 is the length of sbert vector
                coef0 = 1.0
                #polynomial kernel, which is essentially nonlinear cos similarity
                poly_sim = polynomial_kernel(
                    X,
                    degree=degree,
                    gamma=gamma_poly,
                    coef0=coef0
                )[0, 1]


        except (AttributeError):
            print('att')
            sim_list_pearson_cik.append(None)
            sim_list_cos_cik.append(None)
            manhattan_list_cik.append(None)
            poly_list_cik.append(None)
            rbf_list_cik.append(None)
            distance_list_cik.append(None) #if a cik is missing item this will get thrown because we will not be able to do cos similarity_pearson with a null
            continue

        sim_list_pearson_cik.append(similarity_pearson)
        distance_list_cik.append(distance)
        sim_list_cos_cik.append(cos_similarity)
        manhattan_list_cik.append(manhattan)
        poly_list_cik.append(poly_sim)
        rbf_list_cik.append(rbf_sim)


    sim_list_pearson.extend(sim_list_pearson_cik) #add all the similarities for the cik to the overall list
    distance_list.extend(distance_list_cik)
    sim_list_cos.extend(sim_list_cos_cik)
    manhattan_list.extend(manhattan_list_cik)
    rbf_list.extend(rbf_list_cik)
    poly_list.extend(poly_list_cik)
```

Load CRSP (pricing data)

```python
#load CRSP data and clean it
crsp = pl.read_csv(
    CRSP_MONTHLY_W_TICKER_PATH,
    schema_overrides={"NCUSIP": pl.Utf8,"CUSIP":pl.Utf8,"RET":pl.Utf8,"RETX":pl.Utf8} #do this so we can load it in and then clean these cols
)

#chat said to use ncusip instead
crsp = crsp.with_columns(
    pl.col("CUSIP").str.slice(0, 8).alias("cusip8")
)

crsp=crsp.with_columns(pl.col('RET').cast(pl.Float32,strict=False)) #make RET a float
# crsp=crsp.with_columns(pl.col('RETX').cast(pl.Float32,strict=False))

#make date a pl.Date
crsp=crsp.with_columns(pl.col('date').cast(pl.Date))
#filter to 10 and 11 sharecode bc that is standard practice
crsp=crsp.filter((pl.col('SHRCD')==10) |(pl.col('SHRCD')==11))

#rename all the cols to lowercase for consistency
crsp = crsp.rename({c: c.lower() for c in crsp.columns})
#get rid of negative price
crsp=crsp.with_columns((pl.col('prc').abs()).alias('prc'))
crsp
```

```python
#merge CRSP and SEC data
merged=data.join(crsp, on=['date','cusip8'],how='right')
```

Make Quintiles

```python
################################################################
#SHIFT SIGNALS
################################################################

#sort the data
merged=merged.sort(['permno','date'])

#shift signal to avoid lookahead
metrics = ["cos_sim", "p_corr", "euclid", "manhattan", "rbf"]

merged = merged.with_columns([
    pl.col(f"{ITEM}_{m}").shift(1).over("permno").alias(f"{ITEM}_{m}")
    for m in metrics
])

###################################################################
#MAKE BINS
###################################################################

#bin similarity
labels = [str(i) for i in range(5)]

#bin all the metrics
merged = merged.with_columns([
    pl.col(f"{ITEM}_{m}")
      .qcut(5, labels=labels, allow_duplicates=True)
      .over("date")
      .alias(f"{ITEM}_{m}_bin")
    for m in metrics
    if f"{ITEM}_{m}" in merged.columns
])

#####################################################################
#HOLDING
#####################################################################

# sort first
merged = merged.sort(["permno", "date"])

# find all bin columns automatically
bin_cols = [col for col in merged.columns if col.endswith("_bin")]

merged = merged.with_columns([
    pl.col(col)
      .fill_null(strategy="forward", limit=11)
      .over("permno")
      .alias(col)
    for col in bin_cols
])
```

```python
#make port dicts for each similarity calculation

#this is the raw returns of each of them
ports_sim_dict={}

#make the equal weight ports for each sim metric
for sim_metric in metrics:
    ports_sim_dict[sim_metric]=merged.group_by(['date',f"{ITEM}_{sim_metric}_bin"]).agg(pl.col('ret').mean())
    ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].filter(pl.col(f"{ITEM}_{sim_metric}_bin").is_not_null())
    ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].sort('date',f"{ITEM}_{sim_metric}_bin")
    ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].pivot(f"{ITEM}_{sim_metric}_bin",index='date',values='ret')
    if sim_metric in ['cos_sim' , 'p_corr' , 'rbf' ]: #because high values mean very similar
        ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].with_columns((pl.col('4') -pl.col('0')).alias('spread'))
    else: #manhattan and euclid, low values mean similar
        ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].with_columns((pl.col('0') -pl.col('4')).alias('spread'))

sharpes=[]
#caluclate sharpe ratios for each metric
for sim_metric in metrics:
    avg_ret=ports_sim_dict[sim_metric]['spread'].mean() #get the sharpes for each of the spreads
    ret_std=ports_sim_dict[sim_metric]['spread'].std()

    #calcuate sharpe and append to the list
    sharpe=avg_ret/ret_std *np.sqrt(12) #annualize it
    sharpes.append(sharpe)

#make a sharpe df
sharpe_df = pl.DataFrame({
    "metric": metrics,
    "sharpe": sharpes
})
```

```python
#make vol scaled of each port in each port
labels_w_spread=labels+['spread']

for sim_metric in metrics:
    #loop over all the bins in the the data frame
    for bin in labels_w_spread:
        try:
            ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].with_columns(
                (pl.col(bin)/(pl.col(bin).rolling_std(window_size=12).shift(1)*np.sqrt(12))).alias(f'{bin}_vs')
            )
        except Exception:
            continue

#now do cumulative sum
for sim_metric in metrics:
    #loop over all the bins in the the data frame
    for bin in labels_w_spread:
        try:
            ports_sim_dict[sim_metric]=ports_sim_dict[sim_metric].with_columns(
                (pl.col(f'{bin}_vs').cum_sum()).alias(f'{bin}_vs_cum')
                )
        except Exception: #this is because sometimes we will not have all the bins (this happens when we have the same cutoffs)
            continue
```

```python
plt.figure(figsize=(12,6))

#color mapping code courtesy of chat
n = 6
cmap = plt.cm.jet  # or plasma, turbo, etc.
colors = cmap(np.linspace(0, 1, n))

#visualize it
for sim_metric in metrics:
    c=metrics.index(sim_metric)#set the color var

    port_pd=ports_sim_dict[sim_metric].to_pandas() #change to pandas so we can plot with mpl
    x=port_pd['date'] #it should be the same across all the t (roughly the same)

    #loop thru the bins and plot each of the cum rets
    y=port_pd[f'spread_vs_cum']
    plt.plot(x,y,label=f'spread_{sim_metric}',color=colors[c])

plt.legend()
plt.title(f'{ITEM} Spread Across Similarity Metrics (Vol-Scaled, non-log, 11-mo, sbert vec)')
plt.show()
```
